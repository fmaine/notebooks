```python
site_status=[]
for site in sites:
    print(site)
    home_url = 'https://'+site
    r = requests.get(home_url)
    http_status = r.status_code
    html = r.text
    site_status.append({
        'site': site,
        'http_status': http_status,
        'html': html,
    })
```

```python
def get_sitemap_urls(xml):
    urls = []
    soup = BeautifulSoup(xml,"xml")
    s_urls = soup.select("sitemap")
    for s in s_urls:
        urls.append(s.find('loc').text)
    return urls
```

```python
from bs4 import BeautifulSoup
import requests
def get_sitemaps_urls(sitemap_index_url) :
    xml = imp.get_webpage(sitemap_index_url)
    soup = BeautifulSoup(xml,"xml")
    sitemapList = soup.find_all("sitemap")
    allSitemaps = []
    for sitemap in sitemapList:
        allSitemaps.append(sitemap.findNext("loc").text)
    return(allSitemaps)
```

```python
def get_sitemap(site):
    urls=[]
    sitemap_index_url = 'https://'+site+'/sitemap.xml'
    sitemap_urls = get_sitemaps_urls(sitemap_index_url)
    for sitemap_url in sitemap_urls:
        xml = imp.get_webpage(sitemap_url)
        soup = BeautifulSoup(xml,"xml")
        for s_url in soup.select("url"):
            loc=''
            lastmod=''
            s_loc = s_url.find('loc')
            if s_loc:
                loc = s_loc.text
            s_lastmod = s_url.find('lastmod')
            if s_lastmod:
                lastmod = s_lastmod.text
            urls.append({
                'loc': loc,
                'lastmod': lastmod
            })
    return urls
```
