## Countvectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df_label['text']=df_label['text'].apply(str.lower)
```

```python
import re
```

```python
vectorizer = CountVectorizer(lowercase=True,tokenizer=lambda t: re.findall(r'[^\W\d][^\W\d]+',t))
X = vectorizer.fit_transform(df_label['text'])
```

```python
vectorizer.get_feature_names_out()
df_vec = pd.DataFrame(X.toarray(),columns=vectorizer.get_feature_names_out())
```

```python
df_words = pd.DataFrame(df_vec.sum(),columns=['count'])
```

```python
df_words.sort_values('count', ascending=False)
```

```python
stopwords_fr=list(df_words.sort_values('count', ascending=False).index[:1000])
```
