### Fonction de normalisation des labels

```python
import numpy as np
def normalize_label(value):
    if value in ['oui', 'Oui','1', 1.0, 1]:
        return 1.0
    elif value in ['non', 'Non', '0', 0.0, 0]:
        return 0
    else:
        return np.NaN
```

### Normalisation Labels

```python
from patat.imp.labels import label_translator
from patat.imp.labels import short_label_names

df_label = df_label.rename(columns=label_translator)
```

```python
df_label[df_label['infox'].isna()]
```

```python
df_label = df_label[df_label['infox'].notna()]
```

```python
for label in short_label_names:
    if label in df_label.columns:
        df_label[label]=df_label[label].apply(normalize_label)
```

## Parsing articles

```python
import patat.imp.importer
import patat.imp.parser
```

```python
imp = patat.imp.importer.Importer()
parser = patat.imp.parser.Parser()
```

```python
parse_data = []
for url in tqdm(df_label['url']):
    site = imp.get_site(url)
    if site in parser._selectors:
        try:
            html = imp.get_webpage(url)
            parse_data.append( parser.parse_article_selector(url,html,site))
        except:
            print(f'Error : {url}')
    
```

```python
len(parse_data)
```

```python
df_parse = pd.DataFrame(parse_data)
```

```python
df_parse = df_parse.drop_duplicates(subset='url')
```

```python
df_merge = df_label.merge(df_parse,on='url',how='outer')
```

```python
df_merge[df_merge['article'].isna()]
```
