## Extraction paragraphes Recueil

```python
from patat.imp import selectors
from patat.imp.importer import Importer
from patat.imp.parser import Parser
```

```python
importlib.reload(selectors)
```

```python
imp = Importer()
parser = Parser()
```

```python
def get_parag_html(url):
#    print(url)
    site = imp.get_site(url)
    html = imp.get_webpage(url)
    parse_data = parser.parse_article(url,html,site)
    return parse_data.get('paragraphs',[])
```

```python
df_recueil['url'][2]
```

```python
get_parag_html(df_recueil['url'][22])
```

```python
%%time
df_recueil['parag_html']=df_recueil['url'].apply(get_parag_html)
```
