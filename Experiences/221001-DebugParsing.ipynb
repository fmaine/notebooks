## Resynchro sur le cache
---

```python
sites = ['fr.sott.net','www.lelibrepenseur.org','www.francetvinfo.fr']
sites = ['actu.fr']
```

```python
import os
import re
html_files = {}
pages = {}
for site in sites:
    html_files[site] = os.listdir('data/html/'+site)
    pages[site] = []
    for filename in html_files[site]:
        pages[site].append(re.sub(r'\|','/','https://'+site+filename))
```

```python
pages
```

```python
import patat.imp.parser_params
import patat.imp.importer
import patat.imp.parser
```

```python
importlib.reload (patat.imp.parser_params)
importlib.reload (patat.imp.importer)
importlib.reload (patat.imp.parser)
```

```python
imp = patat.imp.importer.Importer()
parser = patat.imp.parser.Parser()
```

```python
i = 1
parsed_articles = []
for site in pages:
    for article_url in pages[site]:
        if parser.is_article_url(article_url,site):
            print(i,article_url)
            html = imp.get_webpage(article_url)
            parsed_articles.append(parser.parse_article(article_url,html,site))
            i = i+1
            if i> 100 : break
```

```python
for item in parsed_articles:
    print(item['title'])
    if '\n' in item['title']:
        print(item['url'])
```
