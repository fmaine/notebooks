## GoogleSheet Recueil

```python
import patat.util.gsheet

gs = patat.util.gsheet.GoogleSheet()

sheet_id = '1RFxwIBYqa_eYcx_LZpTdhi47QbLcesz9NByMfGzdJ8s'
range_name = 'articles labellisés'

df_brut = gs.read_df(sheet_id,range_name)

df_brut
```

## Normalisation labels

```python
from patat.imp.labels import label_translator
from patat.imp.labels import short_label_names

df_recueil = df_brut.rename(columns=label_translator)

import numpy as np
def normalize_label(value):
    if type(value) != str:
        return np.NaN
    elif value == 'oui':
        return 1
    elif value =='non' :
        return 0
    else:
        return np.NaN

for label in short_label_names:
    df_recueil[label]=df_recueil[label].apply(normalize_label)

df_recueil
```

```python
df_recueil['tds']=df_recueil['tds'].astype(int)
```

```python
df_recueil[df_recueil['tds']==0]
```

## Importation et parsing articles

```python
from patat.imp.importer import Importer
from patat.imp.parser import Parser

imp = Importer()
parser = Parser()

df_recueil['site']=df_recueil['url'].apply(imp.get_site)
```

```python
df_recueil
```

### Chargement html

```python
%%time
pages = []
for url in df_recueil['url']:
    print(url)
    try:
        html = imp.get_webpage(url)
        pages.append({
            'url' : url,
            'html' : html,
        })
    except:
        print(f'Error : {url}')

df_pages = pd.DataFrame(pages)

df_recueil = df_recueil.merge(df_pages, on='url')
```

### Parsing de l'ensemble du Recueil

```python
%%time
df_articles = pd.DataFrame(list(df_recueil.progress_apply(lambda row: parser.parse_article(row['url'],row['html'],row['site']), axis=1)))
```

```python
df_articles['c_count']=df_articles['article'].apply(lambda s: len(s) if type(s)==str else 0)
df_articles['p_count']=df_articles['paragraphs'].apply(lambda s: len(s) if type(s)==list else 0)
df_recueil = df_recueil.merge(df_articles,on='url')
```

### Traitement Paragraphes

```python
df_recueil[(df_recueil['p_count']==0)&(df_recueil['tds']==1)]
```

```python
filter_indexes = df_recueil.loc[(df_recueil['p_count']==0)&(df_recueil['tds']==1)].index
df_recueil.loc[filter_indexes,'paragraphs']=df_recueil.loc[filter_indexes,'article'].apply(lambda s: s.split('\n'))
```

### Articles trop longs

```python
c_count_max = 50_000
filter_indexes = df_recueil.loc[(df_recueil['c_count']>c_count_max)&df_recueil['tds']].index
df_recueil.loc[filter_indexes]
```

```python
df_recueil.loc[filter_indexes,'tds']=0
```

```python
df_recueil = df_recueil[df_recueil['c_count']<=c_count_max].copy()
```

### Taille et nombre de paragraphes

```python
df_recueil['p_size']=df_recueil['c_count']/df_recueil['p_count']
```

```python
def print_parags(parags):
    for p in parags:
        print(p,'\n')
```

```python
df_filtre = df_recueil[df_recueil['p_size']>2000].sort_values('p_size')
for url in df_filtre['url']:
    print(url)
```
