```python
df_label = label_db.df()
```

```python
df_label = df_label[(df_label['label']=='infox')&(df_label['owner']=='221028-InfoxPipe')]
```

```python
df_label['site']=df_label['url'].apply(imp.get_site)
```

```python
result = []
for site in df_label['site'].unique():
    df_site = df_label[df_label['site']==site]
    result.append({
        'site' : site,
        'ratio' : df_site[df_site['value']>0.5].shape[0]/df_site.shape[0],
    })
df_ratio = pd.DataFrame(result)
df_ratio.sort_values('ratio',ascending=False)
```

```python
sns.histplot(df_ratio,x='ratio',y='site')
```
