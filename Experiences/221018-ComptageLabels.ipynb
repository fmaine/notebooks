```python
df_label = label_db.df().copy()
```

```python
df_label = df_label[df_label['owner']=='recueil']
```

```python
df_croise = df_label.pivot_table(values='value',index='url',columns='label')
```

```python
df_croise.keys()
```

```python
labels = ['cherry_picking', 'denigrement', 'entites_coherentes',
       'entites_nommees', 'exageration', 'faits', 'fausse_nouvelle',
       'insinuations', 'inverifiable', 'inversion_preuve', 'opinions',
       'ouverture_esprit', 'propos_raportes', 'qualite_ecriture',
       'scientifique_sulfureux', 'signe', 'sophisme', 'sources_citees',
       'titre_decale', 'trop_chiffres', 'trop_mots']
```

```python
df = df_croise[df_croise['infox']==1.0]
print(f'infox = 1 : {len(df)} articles')
print(df[labels].sum())
df = df_croise[df_croise['infox']==0.0]
print(f'infox = 0 : {len(df)} articles')
print(df[labels].sum())
```
