### Fonction de normalisation des labels

```python
import numpy as np
def normalize_label(value):
    if value in ['oui', '1', 1]:
        return 1
    elif value in ['non', '0', 0]:
        return 0
    else:
        return np.NaN
```

### Normalisation Labels

```python
from patat.imp.labels import label_translator
from patat.imp.labels import short_label_names

df_label = df_label.rename(columns=label_translator)
```

```python
df_label = df_label[df_label['infox'].notna()]
```

```python
for label in short_label_names:
    if label in df_label.columns:
        df_label[label]=df_label[label].apply(normalize_label)
```

## Import et parsing des articles

```python
from patat.imp.importer import Importer
from patat.imp.parser import Parser

imp = Importer()
parser = Parser()
```

```python
def get_page(url):
    try:
        return imp.get_webpage(url)
    except:
        print(f'Error : {url}')
        return ''
```

```python
df_label['site'] = df_label['url'].apply(imp.get_site)
```

```python
df_label['html'] = pd.DataFrame(df_label['url'].progress_apply(get_page))
```

```python
df_label['html']
```

```python
parse_data = df_label.progress_apply(lambda row: parser.parse_article(row['url'],row['html'],row['site']), axis=1)
```

```python
df_label = df_label.merge(pd.DataFrame(list(parse_data)))
```

```python
df_label.keys()
```

```python
df_label = df_label[['url','site', 'title', 'article',
       'tags', 'paragraphs', 'date_iso', 'abstract', 'Auteur','author', 'date','infox', 'justification infox',
        'entites_nommees', 'ouverture_esprit', 'faits', 'opinions',
       'propos_raportes', 'sources_citees', 'fausse_nouvelle', 'insinuations',
       'exageration', 'inverifiable', 'titre_decale', 'cherry_picking',
        'remarque observatoire', 'signe',
       'entites_coherentes', 'qualite_ecriture', 'intention',
       'scientifique_sulfureux', 'trop_mots', 'trop_chiffres', 'sophisme',
       'inversion_preuve', 'denigrement', 'degre_infox', 'observateur','source']]
```

## Metriques

```python
df_label[df_label['title'].isna()]
```

```python
df_label[df_label['article'].isna()]
```

```python
def get_p_count(parags):
    if type(parags)==list:
        return len(parags)
    else:
        return 0
```

```python
df_label['p_count'] = df_label['paragraphs'].apply(get_p_count)
```

```python
df_label = df_label[df_label['p_count']!=0]
```

```python
def get_text(row):
    return parser.text_from_data(row['title'],row['abstract'],row['paragraphs'])
```

```python
df_label['text']=df_label.apply(get_text,axis=1)
```

```python
df_label['c_count']=df_label['text'].apply(len)
```

```python
df_label['p_size']=df_label['c_count']/df_label['p_count']
```

```python
import patat.util.text
```

```python
df_label['url_h']=df_label['url'].apply(patat.util.text.compress)
```
