```python
df_fb['fb-infox'] = df_fb['infox'].apply(label_db.normalize_label)
df_fb = df_fb[df_fb['fb-infox'].notna()]
```

```python
df_gp['gp-infox'] = df_gp['infox'].apply(label_db.normalize_label)
df_gp = df_gp[df_gp['gp-infox'].notna()]
```

```python
df_fb['infox']
```

```python
df_gp
```

```python
df_merge = pd.merge(df_fb[['url_h','fb-infox']],df_gp[['url_h','gp-infox']])
```

```python
df_merge
```

```python
df_labels = label_db.df()
```

```python
df_predict = df_labels[df_labels['owner']=='221028-InfoxPipe']
```

```python
df_predict['Id']=df_predict['url_h']
```

```python
df_merge = pd.merge(df_merge,df_predict[['url_h','value']])
```

```python
df_merge['bool_value']=df_merge['value'].apply(lambda v: 1 if v >0.5 else 0)
```

```python
df_merge
```

## Mesures

```python
from sklearn import metrics
```

```python
results_fb_gp = {}
results_fb_gp['f1']=metrics.f1_score(df_merge['fb-infox'], df_merge['gp-infox'])
results_fb_gp['accuracy']=metrics.accuracy_score(df_merge['fb-infox'], df_merge['gp-infox'])
results_fb_gp['recall']=metrics.recall_score(df_merge['fb-infox'], df_merge['gp-infox'])
cnf_matrix = metrics.confusion_matrix(df_merge['fb-infox'], df_merge['gp-infox'],normalize='all')
results_fb_gp['false_positive']=cnf_matrix[0,1]
results_fb_gp['false_negative']=cnf_matrix[1,0]
cnf_matrix
```

```python
results_fb_gp
```

```python
results_gp_fb = {}
results_gp_fb['f1']=metrics.f1_score(df_merge['gp-infox'], df_merge['fb-infox'])
results_gp_fb['accuracy']=metrics.accuracy_score(df_merge['gp-infox'], df_merge['fb-infox'])
results_gp_fb['recall']=metrics.recall_score(df_merge['gp-infox'], df_merge['fb-infox'])
cnf_matrix = metrics.confusion_matrix(df_merge['gp-infox'], df_merge['fb-infox'],normalize='all')
results_gp_fb['false_positive']=cnf_matrix[0,1]
results_gp_fb['false_negative']=cnf_matrix[1,0]
```

```python
results_gp_fb
```
