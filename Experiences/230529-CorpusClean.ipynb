### Reparse pour sites à probleme

```python
import patat.imp.importer
import patat.imp.parser
```

```python
imp = patat.imp.importer.Importer()
parser = patat.imp.parser.Parser()
```

```python
pages = []
for url in tqdm(df_corpus_new[df_corpus_new['title'].isna()]['url']):
    html = imp.get_webpage(url)
    pages.append({
        'url': url,
        'html': html,
    })
```

```python
len(pages)
```

```python
parse_data=[]
for page in tqdm(pages):
    parse_data.append(parser.parse_article_selector(page['url'],page['html'],imp.get_site(page['url'])))
df_reparse = pd.DataFrame(parse_data)
```

## Traitement paragraphes

```python
df_corpus_new['p_count']=df_corpus_new['paragraphs'].apply(len)
```

```python
df_corpus_new[df_corpus_new['p_count']==0].value_counts('site')
```

```python
df_parags = df_corpus_new[df_corpus_new['p_count']==0]
```

```python
def parags_from_article(text):
    result = text.split('\n')
    return [p.strip() for p in result if p.strip() != '']
```

```python
df_parags['paragraphs']=df_parags['article'].progress_apply(parags_from_article)
```

```python
df_parags['paragraphs'].iloc[123]
```

```python
df_corpus_new.loc[df_parags.index,'paragraphs']=df_parags['paragraphs']
```

## Attributs calculés

### text

```python
def make_text(row):
    text = row['title']
    if row['abstract']:
        text += '\n' + row['abstract']
    for p in row['paragraphs']:
        text += '\n' + p
    return text
```

```python
df_corpus_new['text']=df_corpus_new.apply(make_text,axis=1)
```

```python
df_corpus_new['c_count'] = df_corpus_new['text'].apply(len)
```
