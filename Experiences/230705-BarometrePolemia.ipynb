# Chargement et parsing pages

```python
import patat.util.gsheet

gs = patat.util.gsheet.GoogleSheet()

sheet_id = '1VbmqNkN5WZNI6iz2TELicFW6LEYeRdiCnP5eW_pWcyw'
range_name = 'Polemia2022'

df_url = gs.read_df(sheet_id,range_name)
```

```python
df_url
```

```python
import patat.imp.importer
import patat.imp.parser
```

```python
imp = patat.imp.importer.Importer()
parser = patat.imp.parser.Parser()
```

```python
parse_data = []
for index,row in df_url.iterrows():
    url = row['url']
    print(url)
    html = imp.get_webpage(url)
    data = parser.parse_article_selector(url,html,imp.get_site(url))
    data['sm_date_text']=row['date']
    parse_data.append(data)
```

```python
df_site = pd.DataFrame(parse_data)
```

```python
df_site
```

# Prédiction Infox

```python
import patat.util.file

filename = 'data/prod/230705-PredInfox.pkl'

predictor = patat.util.file.pickle_load(filename)
```

```python
df_site['text']=df_site['title']+'\n'+df_site['article']
```

```python
df_site['c_count']=df_site['text'].apply(len)
```

```python
sns.displot(data=df_site,x='c_count')
```

```python
df_site[df_site['c_count']<300]['text'].iloc[0]
```

```python
df_site['proba_infox']=predictor.predict_proba(df_site['text'])[:,1]
```

```python
df_site['pred_infox']=predictor.predict(df_site['text'])
```

```python
df_site['pred_infox'].describe()
```

```python
sns.displot(data=df_site,x='proba_infox',bins=30)
```
