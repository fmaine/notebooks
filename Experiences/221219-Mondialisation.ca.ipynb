```python
import patat.imp.importer
import patat.imp.parser
```

```python
imp = patat.imp.importer.Importer()
parser = patat.imp.parser.Parser()
```

```python
url = 'https://www.mondialisation.ca/'
site = 'www.mondialisation.ca'
```

```python
html = imp.get_webpage(url)
```

```python
data = parser.parse(url,html)
```

```python
article_links = data['article_links']
```

```python
data['article_links']
```

```python
articles = []
for url in data['article_links']:
    print(url)
    html = imp.get_webpage(url)
    articles.append(parser.parse_article(url,html,site))
df_articles = pd.DataFrame(articles)
df_articles
    
```

```python
article_db.merge(df_articles)
```

```python
article_db.df()
```
