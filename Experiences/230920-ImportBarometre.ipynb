## Chargement des site < 10_000 articles

```python
small_sites = ['www.polemia.com','lemediaen442.fr','lesmoutonsrebelles.com','www.mondialisation.ca']
```

```python
df_urls_load = df_urls[df_urls['site'].isin(small_sites)].sample(frac=1,random_state=42)
```

```python
import patat.imp.importer
import patat.imp.parser
```

```python
imp = patat.imp.importer.Importer()
parser = patat.imp.parser.Parser()
```

```python
%%time
article_data = []
for i,url in enumerate(df_urls_load['url']):
    print(f'{i} {url}')
    try:
        html = imp.get_webpage(url)
    except:
        print(f'Error loading {url}')
    if html:
        data = parser.parse_article_selector(url,html,imp.get_site(url))
        article_data.append(data)
    if i%500 == 0:
        print('Saving...')
        df_article = pd.DataFrame(article_data)
        df_article.to_csv('data/tmp/230920-Barometre.csv',index=False)
```

```python
df_article = pd.DataFrame(article_data)
df_article.to_csv('data/tmp/230920-Barometre.csv',index=False)
```
