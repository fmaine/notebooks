```python
from bs4 import BeautifulSoup
```

```python
selector = 'a[rel="tag"]'
```

```python
def get_reltag(html):
    tags = []
    soup = BeautifulSoup(html)
    for s_tag in soup.select('a[rel=tag]'):
        tags.append(s_tag.text.strip())
    return tags
```

```python
get_reltag(df_pages['html'][1])
```

```python
df_pages['reltags']=df_pages['html'].progress_apply(get_reltag)
```
