## Constitution des paragraphes

```python
from patat.imp import selectors
from patat.imp.importer import Importer
from patat.imp.parser import Parser

imp = Importer()
parser = Parser()

def get_parag_html(url):
    site = imp.get_site(url)
    html = imp.get_webpage(url)
    parse_data = parser.parse_article(url,html,site)
    return parse_data.get('paragraphs',[])
```

```python
from tqdm.notebook import tqdm
tqdm.pandas()
```

```python
%%time
df_recueil['parag_html']=df_recueil['url'].progress_apply(get_parag_html)
df_recueil
```

## Prédictions Paragraphes

```python
import patat.util.file
```

```python
classifier = patat.util.file.pickle_load('data/tmp/230425-PredInfox.pickle')
```

```python
def predict_parags(parags):
    if len(parags)>0: 
        return  [row[1] for row in classifier.predict_proba(parags)]
    else:
        return []
```

```python
def predict_parags(parags):
    probas = [row[1] for row in classifier.predict_proba(parags)]
    if len(parags)>0: 
        return  [row[1] for row in classifier.predict_proba(parags)]
    else:
        return []
```

```python
df_recueil['proba_html']=df_recueil['parag_html'].apply(predict_parags)
```

```python
df_recueil['proba_text']=df_recueil['parag_text'].apply(predict_parags)
```

```python
df_recueil['max_proba_text']=df_recueil['proba_text'].apply(lambda probas: max(probas) if probas else None)
```

```python
df_recueil.sort_values('max_proba_text',ascending=False)
```
