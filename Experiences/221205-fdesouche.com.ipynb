## Crawling

```python
import patat.imp.crawler
```

```python
crawler = patat.imp.crawler.Crawler()
```

```python
crawler.crawl_paged_lists('www.fdesouche.com',10,use_cache=True)
```

```python
len(crawler._crawl_map['article_links'])
```

```python
len(crawler._crawl_map['list_links'])
```

```python
articles = []
i = 1
for url in crawler._crawl_map['article_links']:
    print(i,url)
    html = imp.get_webpage(url)
    articles.append(parser.parse(url,html,'www.fdesouche.com'))
    i+=1
```

# Mesure et observation des résultats
---

```python
df_articles = pd.DataFrame(articles)
```
