```python
site = 'www.les-crises.fr'
```

```python
crawler.crawl_paged_lists(site, nb_pages = 10, use_cache = True)
```

```python
len(crawler._crawl_map['article_links'])
```

```python
i=1
articles = []
for url in crawler._crawl_map['article_links']:
    print(f'{i} {url}')
    html = imp.polite_get(url)
    article = parser.parse(url,html,site)
    articles.append(article)
    i = i+1
```

```python
df_site = pd.DataFrame(articles)
```

```python
date_text = df_site['date'][0]
parser.text_to_date(date_text, site)
```

```python
df_site['date_iso'] = df_site['date'].apply(lambda t : parser.text_to_date(t,site))
```

```python
patat.imp.parser_params.date_substring_masks
```

```python
import re
from patat.util.date import month2num
import datetime
```

```python
date_text = patat.util.text.normalize(date_text)
# extract date_data
mask = patat.imp.parser_params.date_substring_masks.get(site,'')
match = re.search(mask,date_text)
if match:
    date_data = match.group(0)
    for month in month2num:
        date_data = re.sub(month,month2num[month],date_data)
    date_format = patat.imp.parser_params.date_formats.get(site,'')
    date = datetime.datetime.strptime(date_data,date_format)
```
