### Comptage Lemmes pour 

```python
def merge_texts(texts):
    result = ''
    for text in texts:
        result = result + text + '\n'
    return result
```

```python
def count_words_label(df,label):
    df = df[df[label].notna()]
    df_texts = pd.pivot_table(df, values='lemmas', index=None, columns=label, aggfunc=merge_texts)
    wc = {}
    word_analysis = {}
    for key in df_texts.keys():
        text = df_texts[key]['lemmas']
        count_colname = label+'_'+str(key)
        wc[key] = lex.count_tokens(lex.get_words(text))
        for word in wc[key]:
            word_dic = word_analysis.get(word,{})
            word_dic[count_colname]=wc[key][word]
            word_analysis[word]=word_dic
    return word_analysis
```

```python
from patat.imp.labels import short_label_names
```

```python
df_counts = []
for label in short_label_names:
    df_count = pd.DataFrame(count_words_label(df_recueil,label)).T
    df_count = df_count.fillna(0)
    df_counts.append(df_count)
```
