### Fonction de normalisation des labels

```python
import numpy as np
def normalize_label(value):
    if value in ['oui', '1', 1.0]:
        return 1
    elif value in ['non', '0', 0.0]:
        return 0
    else:
        return np.NaN
```

### Normalisation Labels

```python
from patat.imp.labels import label_translator
from patat.imp.labels import short_label_names

df_tests = df_tests.rename(columns=label_translator)
```

```python
df_tests[df_tests['infox'].isna()]
```

```python
df_tests = df_tests[df_tests['infox'].notna()]
```

```python
for label in short_label_names:
    if label in df_tests.columns:
        df_tests[label]=df_tests[label].apply(normalize_label)
```

```python
df_tests.reset_index(inplace=True)
```
