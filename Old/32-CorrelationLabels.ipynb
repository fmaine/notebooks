```python
label_columns = ['infox','signé',
       'lieux, dates et personnes ', 'lieux, dates et personnes cohérentes',
       'points de vue différents présentés', 'écriture de qualité',
        'faits', 'opinions', 'fausse nouvelle', 'insinuations',
       #'exagération',
        'scientifique sulfureux', 'des pages et des pages',
       'des chiffres et des tableaux', 'sophisme', 'inversion de la preuve',
       'invérifiable', 'propos rapporté', 'sources citées', 'titre décalé',]
```

```python
for col in label_columns:
    df = df[df[col].notna()]
```

```python
df
```

```python
result = {}
for col1 in label_columns:
    result[col1] = {}
    for col2 in label_columns:
        chi2, p, dof, expected = stats.chi2_contingency(pd.crosstab(df[col1], df[col2]))
        result[col1][col2] = p
pd.options.display.float_format = '{:,.4f}'.format
```

```python
df_pvalue = pd.DataFrame(result)
df_pvalue.round(3).to_csv('data/analysis/220916-PValue.csv',decimal=',')
df_pvalue.round(3)
```

```python
label_columns = ['infox','signé']
```

```python
import numpy as np
from sklearn.metrics import accuracy_score
result = {}
for col1 in label_columns:
    result[col1] = {}
    y_true = list(df[col1])
    for col2 in label_columns:
        y_pred = list(df[col2])
        try:
            accuracy_score = accuracy_score(y_true, y_pred,normalize=True)
        except:
            accuracy_score = None            
        print(col1,col2,accuracy_score)
        result[col1][col2] = accuracy_score
```

```python
col1='infox'
col2='signé'
y_test = list(df[col1])
y_pred = list(df[col2])
match = 0
for i in range (0,len(y_test)):
    if y_pred[i]==y_test[i]:
        match = match+1
print (match/len(y_test))
    
```

```python
df_accuracy = pd.DataFrame(result)
df_accuracy.to_csv('data/analysis/220916-Accuracy.csv',decimal=',')
df_accuracy
```
