On se focalise sur les labels. 
Les autres colonnes ne sont pas utiles.

```python
label_columns = ['infox','signé',
       'lieux, dates et personnes ', 'lieux, dates et personnes cohérentes',
       'points de vue différents présentés', 'écriture de qualité',
        'faits', 'opinions', 'fausse nouvelle', 'insinuations',
       #'exagération',
        'scientifique sulfureux', 'des pages et des pages',
       'des chiffres et des tableaux', 'sophisme', 'inversion de la preuve',
       'invérifiable', 'propos rapporté', 'sources citées', 'titre décalé',]
```

```python
df_labels = df[label_columns].fillna(0)
```

```python
df_labels
```

# Choix des labels

Choisir les labels à comparer : `label_a` et `label_b`


```python
label_a = 'infox'
#label_b = 'signé'
label_b = 'fausse nouvelle'
#label_b = 'sophisme'
```

# Comptages et fréquences/probabilités

```python
print(f"\nComptage du nombre d'occurences pour toutes les combinaisons de {label_a} et {label_b}")
df_cross = pd.crosstab(df_labels[label_a],df_labels[label_b])
df_cross['total']=df_cross[0.0]+df_cross[1.0]
df_cross
```

```python
print(f"\nComptage des {label_a}\n")
nb_occ = df_labels.shape[0]
print(f"Nombre d'occurences : {nb_occ}")
print(f"Nombre de {label_a} = 1 : {df_cross.loc[1,'total']}")
print(f"Nombre de {label_a} = 0 : {df_cross.loc[0,'total']}")
```

```python
print(f"\nFréquences observées pour {label_a}\n")
proba_a1 = df_cross.loc[1,'total']/nb_occ
proba_a0 = df_cross.loc[0,'total']/nb_occ
print(f"Frequence {label_a} = 1 : {proba_a1} = {df_cross.loc[1,'total']}/{nb_occ}")
print(f"Frequence {label_a} = 0 : {proba_a0} = {df_cross.loc[0,'total']}/{nb_occ}")
```

```python
print(f"\nComptage des {label_b}\n")
print(f"Nombre d'occurences : {nb_occ}")
print(f"Nombre de {label_b} = 1 : {df_cross[1].sum()}")
print(f"Nombre de {label_b} = 0 : {df_cross[0].sum()}")
```

```python
print(f"\nProbabilités théoriques pour {label_a}\n")
df_proba = pd.DataFrame([[0,0],[0,0]])
df_proba.loc[0,0] = df_cross[0].sum()*proba_a0
df_proba.loc[1,0] = df_cross[0].sum()*proba_a1
df_proba.loc[0,1] = df_cross[1].sum()*proba_a0
df_proba.loc[1,1] = df_cross[1].sum()*proba_a1
df_proba.round(2)
```

# Calcul de la distance entre l'observation et la théorie

```python
def distance(real,theory):
    return (real - theory)**2/theory
```

```python
df_distance = pd.DataFrame([[0,0],[0,0]])
total_distance = 0
for i in [0,1]:
    for j in [0,1]:
        dist = distance(df_cross.loc[i,j],df_proba.loc[i,j])
        df_distance.loc[i,j]=dist
        total_distance += dist
print(f'\nDistance totale : {total_distance}')
df_distance
```

# Controle avec le Khi 2

```python
from scipy import stats
```

```python
chi2, p, dof, expected = stats.chi2_contingency(df_cross)
print(f'chi2 : {chi2}\np : {p}\ndof : {dof}')
print(expected)
```

```python
if total_distance == chi2:
    print('Hourah !!!\nNous avons calculé le khi2...')
    print(f"c'est {chi2} mais qu'est ce que ca veut dire ?")
```
