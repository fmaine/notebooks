```python
import importlib
import os
import urllib
```

```python
def url_to_directory(url):
    parsed = urllib.parse.urlparse(url)
    directory = parsed.netloc
    return directory
```

```python
def url_to_filename(url):
    parsed = urllib.parse.urlparse(url)
    filename = parsed.path.replace('/', '|')+parsed.query.replace('/', '|')
    if filename == '':
        filename = 'index.html'
    return filename
```

```python
# Renommage des fichiers
def url_from_filename(site,filename):
    url = ''
    base = f'https://{site}'
    if site == 'www.francesoir.fr':
        tokens=filename.split('-')
        url = f'{base}{tokens[0]}/{tokens[1]}-{tokens[2]}/{"-".join(tokens[3:])}'
    return url
```

```python
for filename in os.listdir('data/html-v2/www.francesoir.fr'):
    print(filename)
    old_name = os.path.join('data/html-v2/www.francesoir.fr',filename)
    new_name = os.path.join('data/html-v2/www.francesoir.fr',url_to_filename(url_from_filename('www.francesoir.fr',filename)))
    os.rename(old_name, new_name)
    
```
