### Lemmatization avec http://www.lexique.org/

```python
df_lexique = pd.read_csv('patat/etc/Lexique383.tsv',sep='\t')
```

```python
df_lexique
```

```python
def lemmatize_word(word):
    df_word = df_lexique[df_lexique['ortho']==word]
    if df_word.shape[0] == 0:
        return word
    else:
        return list(df_word['lemme'])[0]
```

```python
lemmatize_word('accompagnera')
```

```python
lemme_dic = {}
for index,row in df_lexique[['ortho','lemme']].iterrows():
    if row['ortho'] not in lemme_dic:
        lemme_dic[row['ortho']] = row['lemme']
len(lemme_dic)
```
