```python
pages = {}
for index in df_corpus[df_corpus['date_iso'].isna()][df_corpus['site']=='www.francesoir.fr'].index:
    url = df_corpus.loc[index,'url']
    pages[url] = imp.get_webpage(url,use_cache=False)
```

```python
i = 0
for url in pages:
    html = pages[url]
    soup = BeautifulSoup(html)
    print(url)
    s = soup.find(class_='field--name-field-date me-3')
    if s:
        date_text = s.text.strip()
        print(date_text)
    i = i+1
    if i>100:
        break
```
