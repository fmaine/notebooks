### Reimport www.lelibrepenseur.org

```python
import patat.imp.crawler
import patat.imp.parser
import patat.imp.importer
```

```python
imp = patat.imp.importer.Importer()
crawler = patat.imp.crawler.Crawler()
parser = patat.imp.parser.Parser()
```

```python
urls = list(df_corpus[df_corpus['site']=='www.lelibrepenseur.org']['url'].values)
```

```python
htmls = {}
for url in urls:
    html = imp.get_webpage(url)
    htmls[url]=html
```

```python
len(htmls)
```

```python
articles = []
i = 1
for url in htmls:
    print(f'{i} {url}')
    article = parser.parse(url,htmls[url],'www.lelibrepenseur.org')
    articles.append(article)
    i = i+1
```

```python
df_reload = pd.DataFrame(articles)
df_reload
```

```python
df_reload[(df_reload.duplicated(['title','site'],keep=False))][['url','title','site']]['url'].values
```

```python
df_reload = df_reload.drop_duplicates(['title','site'])
```

```python
df_reload = df_reload[['url','title','article','date_iso','author']]
```

```python
df_reload['site'] = 'www.lelibrepenseur.org'
```

Remplacement des articles du corpus 

```python
df_corpus = df_corpus[df_corpus['site']!='www.lelibrepenseur.org']
```

```python
df_corpus.value_counts('site')
```

```python
df_corpus = pd.concat([df_corpus,df_reload],ignore_index=True)
```

```python
df_corpus.to_csv('data/corpus/221016-ArticleDedup.csv')
```
