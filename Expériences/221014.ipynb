### Etape 1

```python
import patat.ml.lex_analyser
lex = patat.ml.lex_analyser.LexAnalyser()
```

```python
df_article = article_db.df().copy()
```

```python
df_article['text'] = df_article['title'] + '\n' + df_article['article']
```

```python
df_article['nb_char']=df_article['text'].apply(len)
```

```python
df_article['nb_words']=df_article['text'].apply(lambda text : len(lex.get_words(text)))
```

```python
df_article
```

```python
import seaborn as sns
```

```python
sns.displot(df_article['nb_words'],bins=20)
```

```python
df_article['nb_char'].median()
```

```python
df_article['nb_words'].median()
```

```python
import patat.db.label_db
label_db = patat.db.label_db.LabelDb()
label_db.df()
```

```python
df_label = label_db.df()
df_label_recueil = df_label[(df_label['owner']=='recueil')&(df_label['label']=='infox')].pivot_table(values='value',index='url')
df_label_recueil.reset_index()
df_recueil = pd.merge(df_article,df_label_recueil, on='url')
```

```python
df_recueil = df_recueil.rename(columns = {'value':'infox'})
df_recueil
```

```python
df_recueil['nb_char'].mean()
```

```python
df_recueil[df_recueil['infox']==0]['nb_char'].mean()
```

```python
df_recueil[df_recueil['infox']==1]['nb_char'].mean()
```
